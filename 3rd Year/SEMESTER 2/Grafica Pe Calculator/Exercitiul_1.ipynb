```python
import numpy as np
import moderngl_window as mglw

class App(mglw.WindowConfig):
    gl_version = (3, 3)
    title = "Triangle"
    window_size = (800, 600)

    def __init__(self, **kwargs):
        super().__init__(**kwargs)

        self.prog = self.ctx.program(
            vertex_shader="""
                #version 330
                in vec2 in_pos;
                void main() {
                    gl_Position = vec4(in_pos, 0.0, 1.0);
                }
            """,
            fragment_shader="""
                #version 330
                out vec4 f_color;
                void main() {
                    f_color = vec4(1.0, 0.6, 0.2, 1.0);
                }
            """,
        )

        # IMPORTANT: float32 binary buffer, nu text
        verts = np.array([
            -0.6, -0.5,
             0.6, -0.5,
             0.0,  0.6,
        ], dtype="f4")

        vbo = self.ctx.buffer(verts.tobytes())

        # leagă bufferul de atributul "in_pos"
        self.vao = self.ctx.simple_vertex_array(self.prog, vbo, "in_pos")

    def on_render(self, time: float, frame_time: float):
        self.ctx.clear(0.1, 0.1, 0.12)
        self.vao.render(mode=self.ctx.TRIANGLES)

mglw.run_window_config(App, args=[])
```
