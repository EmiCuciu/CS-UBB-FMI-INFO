```python
import sys
sys.argv = ["moderngl_app"]

import numpy as np
import moderngl_window as mglw

class App(mglw.WindowConfig):
    gl_version = (3, 3)
    title = "Tema 1: Prisma 3D"
    window_size = (800, 600)

    def __init__(self, **kwargs):
        super().__init__(**kwargs)

        self.prog = self.ctx.program(
            vertex_shader="""
                #version 330
                in vec3 in_pos; // Cerința 1: vec3
                uniform float time;
                void main() {
                    float s = sin(time * 0.5);
                    float c = cos(time * 0.5);
                    mat3 rot = mat3(c, 0, s, 0, 1, 0, -s, 0, c);
                    gl_Position = vec4(rot * in_pos * 0.6, 1.0);
                }
            """,
            fragment_shader="""
                #version 330
                out vec4 f_color;
                void main() {
                    f_color = vec4(1.0, 0.6, 0.2, 1.0);
                }
            """,
        )

        # Cerința 4: Coordonate X, Y, Z
        verts = np.array([
            -0.5, -0.5,  0.5,   0.5, -0.5,  0.5,   0.0,  0.5,  0.5, # Baza față
            -0.5, -0.5, -0.5,   0.5, -0.5, -0.5,   0.0,  0.5, -0.5  # Baza spate
        ], dtype="f4")

        # Cerința 3: VBA (Index Buffer)
        indices = np.array([
            0, 1, 2,  3, 5, 4,                   # Bazele
            0, 3, 4,  0, 4, 1,                   # Latura jos
            1, 4, 5,  1, 5, 2,                   # Latura dreapta
            2, 5, 3,  2, 3, 0                    # Latura stanga
        ], dtype="i4")

        self.vbo = self.ctx.buffer(verts.tobytes())
        self.vba = self.ctx.buffer(indices.tobytes())

        # Cerința 5: Modificarea liniei self.vao (format "3f" și index_buffer)
        self.vao = self.ctx.vertex_array(
            self.prog,
            [(self.vbo, "3f", "in_pos")],
            index_buffer=self.vba
        )

    def on_render(self, time: float, frame_time: float):
        self.ctx.clear(0.1, 0.1, 0.12)

        # Cerința 2: Activare wireframe înainte de desenare
        self.ctx.wireframe = True

        if 'time' in self.prog:
            self.prog['time'].value = time

        self.vao.render()

if __name__ == "__main__":
    mglw.run_window_config(App, args=[])
```
